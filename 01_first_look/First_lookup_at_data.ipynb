# First look up at the "House Price" data

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/home/sodbo/work/kaggle/house-prices-advanced-regression-techniques/data/01_raw_data/train.csv')
```

```python
df.head()
```

```python
which_numberic = df.dtypes == 'int64'

df_num = df[which_numberic.index[which_numberic]]
```

```python
X_num = df_num[df_num.columns[0:df_num.columns.size-1]]

Y_num = df_num[df_num.columns[df_num.columns.size-1]]
```

```python
X_num.head()
```

```python
Y_num.head()
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_num, Y_num)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(X_num)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(Y_num, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(Y_num, diabetes_y_pred))
```
